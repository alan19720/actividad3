```python
sns.histplot(data=df , x ="failures" , hue="failures", palette="colorblind")
```

Es de tipo int64 sin embargo analizando los posibles valores, puedo determinar que es una discreta y en distribución la que se acopla más es la "geometrica".

```python
sns.histplot(data=df , x ="G1" , hue="G1", palette="colorblind")
#sns.kdeplot(data=df , x="G1", color="green")
```

A primera vista por ser una variable numerica optaria por que es una distribución continua pero
Analizandola y sabiendo que solo hay 17 posibles valores la considero una distribución discreta de tipo hipergeometrica. 

```python
sns.histplot(data=df , x ="G2" , hue="G2", palette="colorblind")
#sns.kdeplot(data=df , x="G2", color="green")
```

Al igual que la anterior, analizandola y sabiendo que solo hay 16 posibles valores la considero una distribución discreta de tipo hipergeometrica. 

```python
sns.histplot(data=df , x ="G3" , hue="G3", palette="colorblind")
#sns.kdeplot(data=df , x="G2", color="green")
```

# Paso 4

```python
# Crear la nueva columna "GFinal" multiplicando "G3" por 5
df['GFinal'] = df['G3'] * 5
pd.set_option('display.max_rows', None)
# Mostrar el DataFrame resultante
display(df['GFinal'])
```

```python
# Supongamos que tienes un DataFrame llamado df con la columna "GFinal":
total_observaciones = len(df)
conteo_g70_o_mas = len(df[df['GFinal'] >= 70])
probabilidad_g70_o_mas = conteo_g70_o_mas / total_observaciones
print("Probabilidad de obtener un 70 o más", probabilidad_g70_o_mas)
```

```python
conteo_aprobados = len(df[df['GFinal'] >= 60])
probabilidad_aprobados = conteo_aprobados / total_observaciones
print("Probabilidad de aprobar con 60 o más", probabilidad_aprobados)
```

```python
conteo_reprobados = len(df[df['GFinal'] <= 40])
probabilidad_reprobados = conteo_reprobados / total_observaciones
print("Probabilidad de reprobar de manera desastrosa con 40 o menos", probabilidad_reprobados)
```
